```python
# Task 1
import pandas as pd
col_list = ['Date', 'High', 'Low']
df = pd.read_csv('GOOG.csv', usecols = col_list)
df
```

```python
# Task 2
import numpy as np
arr = np.array(df)
arr
```

```python
# Task 3
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(df['Date'], df['High'])
plt.title('Date and High price')
plt.xlabel('Date')
plt.ylabel('High Price')
```

```python
# Task 4
plt.plot(df['Date'], df['Low'], 'r--')
plt.title('Date and Low price')
plt.xlabel('Date')
plt.ylabel('Low price')
plt.show()
```

```python
# Task 5
fig = plt.figure()
ax = fig.add_subplot(111)
ax.fill_between(df['Date'], df['Low'], df['High'])
```
